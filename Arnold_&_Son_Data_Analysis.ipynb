# **Data Analysis**

## **Importing libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Loading the DataFrame**

```python
df = pd.read_csv("/content/RCP S1 G2 Arnold & Son.csv")
```

```python
df
```

## **Data Exploration**

```python
# Number of collections
num_collections = df['parent_model'].nunique()
print(f"\nNumber of collections: {num_collections}")
```

```python
# Count of watches in each collection
collection_counts = df['parent_model'].value_counts()
print("\nCount of watches in each collection:")
print(collection_counts)
```

```python
# Calculate the total number of watches based on Reference_number
total_watches = df['reference_number'].nunique()

# Print the total number of watches
print(f"Total number of watches: {total_watches}")
```

```python
# Filter the DataFrame to include only limited edition watches
limited_edition_watches = df[df['marketing_name'].notnull()]

# Count the number of limited edition watches
total_limited_edition_watches = len(limited_edition_watches)

# Print the total number of limited edition watches
print(f"Total number of limited edition watches: {total_limited_edition_watches}")
```

## **Data Visualization**

```python
# Count of watches in each collection
plt.figure(figsize=(10, 6))
sns.countplot(y='parent_model', data=df, order=df['parent_model'].value_counts().index)
plt.title('Count of Watches in Each Collection')
plt.xlabel('Count')
plt.ylabel('Collection')
plt.show()
```
